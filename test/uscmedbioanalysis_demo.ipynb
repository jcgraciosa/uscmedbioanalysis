## Workflows supported:
As of version 0.1, the following methods have been implemented:
1. read_msd_files - reads the particle MSD data present in the input directory
2. plot_single_msd - for visualizing the individual MSD (use this plot to decide the time interval of interest)
3. plot_ensemble_msd - for visualizing the ensemble MSD (again, you can use this plot to decide the time interval of interest)
4. get_diffusivity_single - perform fitting to obtain the diffusivity for each particle
5. get_diffusivity_ensemble - perform fitting to obtain the diffusivity for the ensemble
6. plot_diffusivity_dist - after running get_diffusivity_single, you can run this method to get a distribution of the diffusivity values

```python
import os
import pandas as pd
import uscmedbioanalysis as medbio
% matplotlib inline
```

### Initialization
The package needs two inputs during initialization: 
1. indir: directory containing the individual MSD files
2. outdir: directory where the output files and plots are written

```python
indir = 'D:\\USC Medical Biophysics\\data\\vm\\s1s2_msd'
#indir = 'D:\\USC Medical Biophysics\\uscmedbioanalysis\\test\\test_track'
outdir = 'D:\\USC Medical Biophysics\\uscmedbioanalysis\\test\\test_out'
analyzer = medbio.TrackAnalyzer(indir, outdir)
```

```python
analyzer.read_msd_files(save_collated = True, show_msd = False)
```

```python
analyzer.msd_data
```

## Working with individual particles

```python
analyzer.plot_single_msd(save_plot = True, plot_suffix = '1')
```

```python
analyzer.get_diffusivity_single(save_param = True)
analyzer.plot_diffusivity_dist()
```

## Working with the ensemble

```python
analyzer.plot_ensemble_msd(save_plot = True, plot_suffix = '1')
```

```python
analyzer.get_diffusivity_ensemble()
```
